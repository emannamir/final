```python
import matplotlib.pyplot as plt
#making a line graph
x_values = [10, 15, 20, 25, 30]
y_values = [19, 20, 21, 22, 23]

fig, ax = plt.subplots()
ax.plot(x_values, y_values)

ax.set_xlabel('age')
ax.set_ylabel('year')
ax.set_title('Vape increase')
ax.grid(True)

plt.show()
```

```python
import matplotlib.pyplot as plt
#making an area graph
x_values = [10, 12, 14, 16, 18]
y_values1 = [2017, 2018, 2019, 2020, 2021]
y_values2 = [1, 3, 5, 7, 9]

fig, ax = plt.subplots()
ax.fill_between(x_values, y_values1, y_values2)

ax.set_xlabel('Age')
ax.set_ylabel('Year')
ax.set_title('Vape epidemic')
ax.grid(True)

plt.show()
```

```python
#building a 3d scatter plot

import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
#starting with importing 
#https://plotly.com/python/3d-scatter-plots/
#https://chat.openai.com/?model=text-davinci-002-render


x_values = [1,2,3,4,5,6]
y_values = [4,5,6,7,8,9]
z_values = [2,4,6,7,8,10]

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.scatter(x_values, y_values, z_values)

ax.set_xlabel('year')
ax.set_ylabel('age')
ax.set_zlabel('Juul sales')

plt.show()
```
